```python

import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

#loading my data

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

#data clean up

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
#Giving a name for our index label
df.index.name = 'job_index'
```

```python
# filtering job by country for country Ireland
df_ireland = df[df['job_country']== 'Ireland' ]

#Sorting out the indexes of our data to be in order 
df_ireland.reset_index(inplace=True)
df_ireland

#droping index 

df_ireland.set_index('job_index', inplace= True)
df_ireland
```

```python
#pivotting  data for salary_year_avg column with the index for job_title-short to get the median, min and max for df_ireland 
median_pivot = df_ireland.pivot_table(values='salary_year_avg', index='job_title_short', aggfunc=['median', 'min', 'max'])

median_pivot

```

```python
#dorting our data by median for salary_year_avg inside a tuple in deascending order
median_pivot.sort_values(by=('median', 'salary_year_avg'), ascending= False)
```

```python
#sorting in pace
median_pivot.sort_index(inplace=True)
median_pivot
```
