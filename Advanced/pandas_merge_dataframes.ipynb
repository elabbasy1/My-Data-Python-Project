```python

import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

#loading my data

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

#data clean up

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
#importing data 
df_usa_software_pivot = pd.read_csv('https://lukeb.co/software_csv', index_col= 'job_posted_month')

#merging our data on the job_posted_month 
df_us_merged = df_usa_software_pivot.merge(df_usa_software_pivot, on='job_posted_month')

# calculating the sum of our data, not in ascending order for the top 5, inclduing the index and converting it to a list

top_5 = (df_us_merged
         .sum().sort_values(ascending= False) # sum and not in ascending order
         .head() #top 5 
         .index # getting the index 
         .to_list()) # converting it to a list

top_5


```

```python

# ploting our data using a line graph
df_us_merged[top_5].plot(kind='line')

plt.title('Monthly job postings for top tech job ') # title
plt.xlabel('2023') # xlabel is 2023
plt.ylabel('job_count') # ylabel is job_count
plt.ylim(0, 20000) # shiffting it up to 20000
plt.legend()
plt.show
```
