```python

import ast
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

#loading my data

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

#data clean up

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])

#using the apply method with lambda to get data if not nall values 

df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df['job_work_from_home'].value_counts().plot(kind = 'pie', startangle = 90, autopct = '%1.1f%%') #ploting data for the selectedf column as a piechart 
plt.title('Work fromhome') #setting title
plt.ylabel('') # no ylable
plt.show()
```

```python
#setting the plots for our 3 piecharts
fig, ax = plt.subplots(1,3)
#creating a dict for the columns we need
dict_column = {            
    'job_work_from_home': 'Work from home',
    'job_no_degree_mention': 'Job degree req',
    'job_health_insurance': 'Health insurance offered'
}
#we will use a for loop withe the enumerate method to be able to have two values, column and title from the dic we created
for i, (column, title) in enumerate(dict_column.items()):
    ax[i].pie(df[column].value_counts(), startangle = 90, autopct = '%1.1f%%') # creating the piecharts for every column and labling the percentage
    ax[i].set_title(title) #setting up the titles
    


plt.show()
```
